```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generar datos
np.random.seed(42)
data = {
    'ID_Transaccion': range(1001, 1051),
    'Canal': np.random.choice(['Web Desktop', 'App Móvil'], 50, p=[0.4, 0.6]),
    'Metodo_Pago': np.random.choice(['Tarjeta Crédito', 'PayPal', 'Transferencia'], 50),
    'Total_Venta': np.random.randint(20, 500, 50),
    'Categoria': np.random.choice(['Electrónica', 'Moda', 'Hogar'], 50)
}

df = pd.DataFrame(data)
print("====================================")
print("Datos Generados")
print("====================================\n\n")
print(df)

print("\n\n-----PARTE 1-----\n\n")

## KPI 1: Ingreso Total por Canal
## Calcular
ingreso_por_canal = df.groupby('Canal')['Total_Venta'].sum()
print("====================================")
print("KPI 1: Ingreso Total por Canal")
print("====================================")
print("Ingreso Total por Canal:")
print(ingreso_por_canal)

# Gráfico
plt.bar(ingreso_por_canal.index, ingreso_por_canal.values, color=['red', 'green'])
plt.title('Ingreso Total por Canal')
plt.ylabel('USD')
plt.show()

## KPI 2: Ticket Promedio por Canal
# Calcular
ticket_promedio = df.groupby('Canal')['Total_Venta'].mean()
print("\n====================================")
print("KPI 2: Ticket Promedio por Canal")
print("====================================")
print("Ticket Promedio por Canal:")
print(ticket_promedio)

# Gráfico
plt.bar(ticket_promedio.index, ticket_promedio.values, color=['red', 'green'])
plt.title('Ticket Promedio por Canal')
plt.ylabel('USD')
plt.show()

## KPI 3: Volumen de Transacciones
# Calcular
transacciones = df['Canal'].value_counts()
print("\n====================================")
print("KPI 3: Volumen de Transacciones")
print("====================================")
print("Volumen de Transacciones:")
print(transacciones)

# Gráfico de pastel
plt.pie(transacciones.values, labels=transacciones.index, autopct='%1.1f%%')
plt.title('Distribución de Transacciones')
plt.show()

## KPI 4: Ingreso por Categoría y Canal
# Calcular
ingreso_categoria = df.groupby(['Canal', 'Categoria'])['Total_Venta'].mean().unstack()
print("\n====================================")
print("KPI 4: Ingreso por Categoría y Canal")
print("====================================")
print("Ingreso Promedio por Categoría y Canal:")
print(ingreso_categoria)

# Gráfico
ingreso_categoria.plot(kind='bar')
plt.title('Ingreso Promedio por Categoría y Canal')
plt.ylabel('USD')
plt.xticks(rotation=0)
plt.show()

print("\n-----PARTE 2-----\n")

# Cargar datos
data = pd.read_csv('ecommerce.zip')

## 1. KPI de logistica

## KPI 1.1: Costo por Modo de Envío
# Calcular
costo_envio = data.groupby('Mode_of_Shipment')['Cost_of_the_Product'].mean()
print("====================================")
print("KPI 1.1: Costo por Modo de Envío")
print("====================================")
print("Costo Promedio por Envío:")
print(costo_envio)

# Gráfico
costo_envio.plot(kind='bar')
plt.title('Costo Promedio por Modo de Envío')
plt.ylabel('USD')
plt.show()

## KPI 1.2: Cantidad de Productos por Categoría
# Calcular
ocupacion = data['Warehouse_block'].value_counts()
print("\n============================================")
print("KPI 1.2: Cantidad de Productos por Categoría")
print("============================================")
print("Ocupación por Bloque:")
print(ocupacion)

# Gráfico
ocupacion.plot(kind='bar')
plt.title('Productos por Bloque de Bodega')
plt.ylabel('Cantidad')
plt.show()

## KPI 1.3: Peso por Modo de Envío
# Calcular
peso_envio = data.groupby('Mode_of_Shipment')['Weight_in_gms'].mean()
print("\n====================================")
print("KPI 1.3: Peso por Modo de Envío")
print("====================================")
print("Peso Promedio por Envío:")
print(peso_envio)

# Gráfico
peso_envio.plot(kind='bar')
plt.title('Peso Promedio por Modo de Envío')
plt.ylabel('Gramos')
plt.show()

## 2. KPI Comerciales

## KPI 2.1: Ingreso Total y Costo Promedio
# Calcular
ingreso_total = data['Cost_of_the_Product'].sum()
costo_prom = data['Cost_of_the_Product'].mean()
print("\n========================================")
print("KPI 2.1: Ingreso Total y Costo Promedio")
print("========================================")
print(f"Ingreso Total: ${ingreso_total:,.2f}")
print(f"Costo Promedio: ${costo_prom:.2f}")

# Gráfico
plt.hist(data['Cost_of_the_Product'], bins=20, color='blue', alpha=0.7)
plt.axvline(costo_prom, color='red', label=f'Promedio: ${costo_prom:.2f}')
plt.title('Distribución de Costos')
plt.xlabel('USD')
plt.legend()
plt.show()

## KPI 2.2: Descuento Promedio
#Calcular
descuento_prom = data['Discount_offered'].mean()
print("\n====================================")
print("KPI 2.2: Descuento Promedio")
print("====================================")
print(f"Descuento Promedio: {descuento_prom:.2f}%")

# Gráfico
data['Discount_offered'].hist(bins=20, color='orange', alpha=0.7)
plt.axvline(descuento_prom, color='red', label=f'Promedio: {descuento_prom:.2f}%')
plt.title('Distribución de Descuentos')
plt.xlabel('% Descuento')
plt.legend()
plt.show()

## KPI 2.3: Margen por Importancia
# Calcular margen (precio = costo * 1.3)
data['Margen'] = data['Cost_of_the_Product'] * 0.3
margen_importancia = data.groupby('Product_importance')['Margen'].mean()
print("\n====================================")
print("KPI 2.3: Margen por Importancia")
print("====================================")
print("Margen Promedio por Importancia:")
print(margen_importancia)

# Gráfico
margen_importancia.plot(kind='bar')
plt.title('Margen por Importancia del Producto')
plt.ylabel('USD')
plt.show()

## 3. KPI de Clientes

## KPI 3.1: CSAT Promedio
# Calcular
csat = data['Customer_rating'].mean()
print("\n====================================")
print("KPI 3.1: CSAT Promedio")
print("====================================")
print(f"CSAT Promedio: {csat:.2f}/5.0")

# Gráfico
data['Customer_rating'].value_counts().sort_index().plot(kind='bar')
plt.axhline(y=len(data)/5, color='red', linestyle='--', label='Promedio')
plt.title('Distribución de Calificaciones')
plt.xlabel('Estrellas')
plt.ylabel('Clientes')
plt.legend()
plt.show()

## KPI 3.2: Retención de Clientes
# Calcular
nuevos = data[data['Prior_purchases'] <= 1].shape[0]
recurrentes = data[data['Prior_purchases'] >= 2].shape[0]
total = len(data)
print("\n====================================")
print("KPI 3.2: Retención de Clientes")
print("====================================")
print(f"Clientes Nuevos: {nuevos/total*100:.1f}%")
print(f"Clientes Recurrentes: {recurrentes/total*100:.1f}%")

# Gráfico
plt.bar(['Nuevos', 'Recurrentes'], [nuevos, recurrentes], color=['red', 'green'])
plt.title('Clientes Nuevos vs Recurrentes')
plt.ylabel('Cantidad')
plt.show()

## KPI 3.3: Llamadas al Soporte
# Calcular
llamadas = data['Customer_care_calls'].mean()
print("\n====================================")
print("KPI 3.3: Llamadas al Soporte")
print("====================================")
print(f"Promedio de Llamadas: {llamadas:.2f}")

# Gráfico
data['Customer_care_calls'].value_counts().sort_index().plot(kind='bar')
plt.axhline(y=len(data)/5, color='red', linestyle='--', label='Promedio')
plt.title('Distribución de Llamadas al Soporte')
plt.xlabel('Número de Llamadas')
plt.ylabel('Clientes')
plt.legend()
plt.show()
```

### 3. Experiencia del Cliente

- El CSAT promedio es 2.97 / 5.0, indicando una satisfaccion media-baja (39.7% califica con 1 o 2 estrellas)
- El 78.9% de los clientes son recurrentes o frecuentes, indicando una buena retencion
- El promedio de llamadas al soporte es 4.02, con el 53.7% haciendo entre 3 y 4 llamadas
